```python
# finding Distribution of Ages using histogram using Seaborn Visualization
plt.figure(figsize=(8, 6))
sns.histplot(df['AGE'], bins=10, kde=True)
plt.title('Distribution of Age')
plt.show()
```

```python
# Gender Counts using bar chart
gc=df['GENDER'].value_counts().reset_index()
gc.columns=['gender','Counts']

plt.figure(figsize=(6,4))
fig, ax = plt.subplots()
var = ax.bar(data = gc,x='gender',height='Counts')
ax.bar_label(var)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Counts')
plt.show()
```

```python
# Lung_cancer Counts using BAR plot
cc=df['LUNG_CANCER'].value_counts().reset_index()
cc.columns=['cancer','Counts']

plt.figure(figsize=(6,4))
fig, ax = plt.subplots()
var = ax.bar(data = cc,x='cancer',height='Counts')
ax.bar_label(var)
plt.title('Cancer & non-cancer counts')
plt.xlabel('Gender')
plt.ylabel('Counts')
plt.show()
```

```python
# Gender Counts for Priliminary analysis
c = df[df['LUNG_CANCER'] == 'YES']
c = c['GENDER'].value_counts().reset_index()
c.columns = ['gender','counts']
print(c)
```

```python
# Cancer Distribution Among Genders

plt.figure(figsize=(6,4))
fig, ax = plt.subplots()
var = ax.bar(c['gender'],c['counts'])
ax.bar_label(var)
plt.xlabel('Gender')
plt.ylabel('counts')
plt.title('Cancer Distribution')
plt.show()
```

```python
# Cancer Distribution Among Smokers (1) and Non-Smokers (0)
sns.countplot(x='SMOKING', hue='LUNG_CANCER', data=df)
plt.title('Smoking vs Lung Cancer')
plt.show()
```

```python

# Create a contingency table
ct = pd.crosstab(df['SMOKING'], df['LUNG_CANCER'], margins=True).reset_index()
print(ct)
```

```python
# Non-smokers percentage among All
ns = ct[ct['SMOKING'] == 0]
ns =ns['YES']* 100 / ns['All']
print(ns)
```

```python
# Smokers percentage among All
sm = ct[ct['SMOKING'] == 1]
sm =sm['YES']* 100 / sm['All']
print(sm)
```

```python
# Create a contingency table
ct = pd.crosstab(df['ALCOHOL_CONSUMING'], df['LUNG_CANCER'], margins=True).reset_index()
print(ct)
```

```python
# Non-drinkers percentage among All
ns = ct[ct['ALCOHOL_CONSUMING'] == 0]
ns =ns['YES']* 100 / ns['All']
print(ns)
```

```python
# Drinkers percentage among All
sm = ct[ct['ALCOHOL_CONSUMING'] == 1]
sm =sm['YES']* 100 / sm['All']
print(sm)
```

```python
# Create a contingency table
ct = pd.crosstab( df['LUNG_CANCER'],df['GENDER'],margins=True).reset_index()
print(ct)
```

```python
# Getting data of Lung_cancer People
ca = ct[ct['LUNG_CANCER'] == 'YES']
print(ca)
```

```python
# Male Percentage among Lung_cancer People
ms = ca['M']*100 / ca['All']
print(ms)
```

```python
# Female Percentage among Lung_cancer People
fs = ca['F']*100 / ca['All']
print(fs)
```

```python
# Filter the data for patients with lung cancer
c = df[df['LUNG_CANCER'] == 'YES']

# Count the occurrences of each gender
gender_counts = c['GENDER'].value_counts()

# Create a pie chart
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%',startangle = 90)
plt.title('Gender Distribution Among Lung Cancer Patients')
plt.show()
```

```python
# Drawing pairplot for all parameters that might causes cancer
sns.pairplot(df, hue='LUNG_CANCER')
plt.show()
```

```python
# Created Box plot to identify outliers 
sns.boxplot(y=df['AGE'])
plt.title('Box Plot of Age')
plt.show()
```
